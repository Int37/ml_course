11) Среди кого больше доля зарабатывающих много (>50K): среди женатых или холостых мужчин (признак marital-status)? Женатыми считаем тех, у кого marital-status начинается с Married (Married-civ-spouse, Married-spouse-absent или Married-AF-spouse), остальных считаем холостыми.

```python
a = dict(df[df['income'] == ' >50K']['marital.status'].value_counts().items())
summarried = 0
sumnonmarried =0 
summarried2 =0
sumnonmarried2 = 0

for i in a:
    if 'Married' in i:
        summarried += a[i]
    else:
        sumnonmarried += a[i]
c = dict(df['marital.status'].value_counts().items())
for i in c:
    if 'Married' in i:
        summarried2 += c[i]
    else:
        sumnonmarried2 += c[i]
if(summarried/summarried2 > sumnonmarried/sumnonmarried2):
    print('Доля зарабатывающих много больше среди женатых')
else:
    print('Доля зарабатывающих много больше среди холостых')
    
```
